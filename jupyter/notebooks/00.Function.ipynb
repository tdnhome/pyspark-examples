## Variable-Length Arguments

Variable-length arguments allow a function to accept any number of arguments using two special syntaxes: 
- *args for non-keyword arguments (list)
- **kwargs for keyword arguments  (dictionary)

````python
def sum_all(*args):
    """Calculates the sum of a variable number of arguments."""
    total = 0
    for num in args:
        total += num
    return total

print(sum_all(1, 2, 3))
print(sum_all(10, 20, 30, 40))
````

````python
def display_info(**kwargs):
    """Displays information passed as keyword arguments."""
    for key, value in kwargs.items():
        print(f"{key}: {value}")

display_info(name="John", age=25, city="Exampleville")
````

## Nested Functions

A nested function (or inner function) in Python is a function defined inside another function. This structure provides encapsulation, allowing the inner function to access variables from the outer function's scope while being hidden from the global scope.

Key Characteristics
- Encapsulation: The inner function is local to the outer function and cannot be accessed from outside it, which hides helper logic and prevents namespace pollution.
- Access to Outer Scope: A key feature is that the inner function can access variables and parameters of the enclosing (outer) function, known as a "nested scope".
- Code Organization: They help organize code by grouping related functionality. If a helper function is only used by one other function, it makes sense to nest it to show it belongs exclusively to that function's task. 

Common Use Cases
- Helper Functions: They can serve as small, specific helper functions for a more complex outer function, improving readability and reducing code duplication within the parent function.
- Closures: This advanced technique involves the outer function returning the inner function without calling it. The inner function retains access to the outer function's variables even after the outer function has finished executing, which allows it to preserve state across calls.
- Decorators: Nested functions, specifically closures, are the primary mechanism for implementing Python decorators. Decorators are used to modify the behavior of another function or class dynamically.


````python
def outer_function(name):
    # variable in the outer function's scope
    greeting = "Hi, " 

    def inner_function():
        # inner function can access the outer function's variables
        print(greeting + name + "!") 

    # the outer function calls the inner function
    inner_function()

# Calling the outer function
outer_function("Larry Nguyen")
````

## Closures (Functions that carry states)

A closure in Python is a nested (inner) function that retains access to variables from its enclosing scope even after the outer function has finished executing. This allows the inner function to "remember" and use data that would normally be deleted after the outer function completes, effectively allowing a function to carry state with it.

To create a closure in Python, three conditions must be met: 
- There must be a nested function (a function defined inside another function).
- The nested function must reference a variable from its enclosing (outer) function's scope. This is a "free variable" to the inner function.
- The enclosing function must return the nested function. 



````python
def make_multiplier(seed):
    # 'seed' is a variable in the outer function's local scope
    def multiply(y):
        # 'inner_function' accesses 'seed' from the enclosing scope
        return seed * y
    # The outer function returns the inner function object
    return multiply 

times3 = make_multiplier(3) # seed = 3 is remembered (carry state) even after outer finishes. 
times3(5)  # 15 
````

## Lambda Functions (Anonymous)

A lambda function in Python is a small, anonymous function defined using the lambda keyword that can take any number of arguments but is limited to a single expression. The result of this single expression is automatically returned without needing a return statement.

Key Characteristics
- Anonymous: It does not require a name, unlike functions defined with the def keyword.
- Single Expression: The body of a lambda function can only contain one expression. It cannot include multiple statements, loops, or complex control flow (like try...except or if/else statements, though conditional expressions can be used).
- Concise Syntax: Lambdas provide a compact way to define simple functions in a single line of code.
- No Return Keyword: The result of the expression is implicitly returned. 

The basic syntax is: lambda arguments: expression
- lambda: The keyword that marks the beginning of the lambda function.
- arguments: Inputs passed to the function, separated by commas.
- expression: The single operation performed on the arguments, the result of which is returned.

When to Use Lambda Functions
Lambda functions are most powerful when used as simple, one-time-use functions, often as arguments to higher-order functions. Common use cases include: 
- map(): Applying a function to every item in an iterable.
- filter(): Selecting elements from an iterable that satisfy a condition.
- sorted(): Providing a custom key for sorting a list or other iterable.
- List Comprehension: Will see in the next chapter
- Data Manipulation: In libraries like pandas, they are often used with methods like .apply() for quick data transformations.
 

````python
doubler = lambda x: x * 2  # Define a lambda and assign a variable to it
print(doubler(5))          # Call the lambda
````

````python
students = [("Emil", 25), ("Tobias", 22), ("Linus", 28)]
# Sort the list of tuples by the second element (age)
sorted_students = sorted(students, key=lambda x: x[1])
print(sorted_students)
````

## Recursion (Recursive Function)

A recursive function in Python is a function that calls itself to solve a problem by breaking it down into smaller, simpler subproblems. Every recursive function must have two main components to function correctly and prevent infinite loops: a base case and a recursive case. 

Key Components
- Base Case: This is the stopping condition. It is a simple case for which the solution is known, and it tells the function when to stop calling itself. Without a base case, the function would run indefinitely, leading to a RecursionError: maximum recursion depth exceeded.
- Recursive Case: This is the part of the function that calls itself with a modified, typically smaller, argument. Each recursive call works on a smaller piece of the original problem, moving closer to the base case.

````python
def factorial(n):
    if n == 1:
        # Base case: when n is 1, stop recursing
        return 1
    else:
        # Recursive case: call the function again with a smaller argument
        return n * factorial(n - 1)

num = 4
print("The factorial of", num, "is", factorial(num))

num = 5
print("The factorial of", num, "is", factorial(num))
````

## Function Annotations (Typing)

Function annotations in Python are a syntax for attaching metadata to function parameters and return values. Since the introduction of PEP 484, their primary use has been for providing type hints, which indicate the expected data types for a function's inputs and the type of value it will return. 

Key Concepts
- Documentation and Readability: Type hints act as built-in documentation, making the code easier for human developers to read, understand, and maintain, especially in large projects or team collaborations.
- Static Type Checking: The Python interpreter itself does not enforce these types at runtime; Python remains a dynamically typed language. Instead, third-party tools called static type checkers, such as Mypy or Pyright, can analyze the code to flag potential type inconsistencies or errors before the program is run.
- IDE Support: Modern Integrated Development Environments (IDEs) and linters use type annotations to provide better features like code completion, argument suggestions, and early error warnings, which improves development speed and code quality.
- Metadata Storage: Python stores these annotations in a special dictionary attribute called __annotations__ on the function object, making them accessible to other libraries and tools for various purposes like validation or API generation (e.g., FastAPI)

Syntax and Examples
- Function annotations use a colon (:) after the parameter name for arguments and an arrow (->) before the final colon of the function definition for the return value.


````python
def greet(name):
    """Returns a greeting message given a name."""
    return f"Hello, {name}"

# Update the above function
# name: str: Indicates the name parameter is expected to be a string.
# -> str: Indicates the function is expected to return a string. 
def greet(name: str) -> str:
    """Returns a greeting message given a name."""
    return f"Hello, {name}"

print(greet("Trung"))
````

````python
from typing import List, Optional

# The pipe operator (|) (available in Python 3.10+) is used to specify a union of types, 
# meaning the value can be of either type. Optional[int] (or int | None in Python 3.10+) 
# signifies that an argument can be an integer or None
def add_em_up(nums: List[int | float]) -> int | float:
    """Returns the sum of a list of integers or floats."""
    total = 0
    for num in nums:
        total += num
    return total

print(add_em_up([1,2,3]))
print(add_em_up([1.1,2.2,3.3]))
````

## Docstring (documentation string)

A docstring in Python is a string literal that serves as a built-in help text for a module, function, class, or method. It is a core part of Python best practices for creating readable and maintainable code. 

Key Characteristics
- Placement: A docstring must appear as the very first statement within the definition of the object it documents (module, function, class, or method).
- Syntax: Docstrings are enclosed in triple quotes (either """ or '''), which allows them to span multiple lines.
- Accessibility: Unlike comments (which are ignored by the interpreter), docstrings are retained at runtime and stored in the object's special __doc__ attribute.
- Purpose: They describe the purpose and usage of the code (what it does, its parameters, return values, and potential exceptions), rather than the how of the implementation.
- Accessing: Users can access docstrings using the built-in help() function (e.g., help(my_function)) or by directly accessing the __doc__ attribute (e.g., my_function.__doc__). 



````python
def add_numbers(a, b):
    """
    Return the sum of two decimal numbers in binary digits.

    Parameters:
        a (int): A decimal integer
        b (int): Another decimal integer

    Returns:
        binary_sum (str): Binary string of the sum of a and b
    """
    binary_sum = bin(a+b)[2:]
    return binary_sum

help(add_numbers)
````

## Function As Class's Method

In Python, "functions as methods" refers to the concept that methods are fundamentally functions that are associated with an object or a class. When a function is defined inside a class, it is called a method and gains the ability to operate on the data (attributes) of that specific object instance.

Key Distinctions
The main difference between a standalone function and a method is their relationship to objects: 
1. Definition Location:
- A function is defined independently, outside of any class, and is called by its name (e.g., len(my_list)).
- method is defined within a class definition.
2. Invocation Syntax:
- Functions are called directly: function_name(arguments).
- Methods are called using dot notation on an object instance: object.method_name(arguments).
3. Implicit First Parameter (self):
- When a method is called using dot notation (e.g., my_dog.bark()), Python automatically passes the instance object itself as the first argument to the method. This parameter is conventionally named self.
- A standalone function has no such implicit parameter.

How It Works.

Essentially, a method is a function that receives a reference to the object it's called on. This allows the method to access and modify the object's state (its internal variables). 


````python
# Define a class with methods. Note that each method received the instance object self
class Dog:
    # The constructor
    def __init__(self, name):  # The constructor
        self.name = name # An attribute (state)

     # A method (behavior)
    def bark(self):           
        # It can access the object's attribute via 'self.name'
        return f"Hey {self.name}, says woof!"

# Create an instance of Dog and call method on an object instance
my_dog = Dog("Rex")
print(my_dog.bark()) 
````

## Special Function Type: Generator Functions (yield)

In Python, a generator function is a special type of function that uses the yield keyword instead of return to produce a sequence of values one at a time. 

Unlike a normal function that returns a single value and terminates completely, a generator function pauses its execution and saves its state (local variables, instruction pointer, etc.) each time it encounters a yield statement. When the generator is called upon again (e.g., in a for loop or by calling next()), it resumes execution from exactly where it left off. 

Key Characteristics and Benefits

- Memory Efficiency: Generators are "lazy iterables"; they generate values on demand and do not store the entire sequence in memory. This makes them ideal for processing large datasets or files that wouldn't fit in memory otherwise.
- State Retention: The function's state is automatically saved between calls, allowing it to continue from the exact point it paused.
- Infinite Sequences: Since values are produced one at a time, generators can be used to represent infinite sequences (like the Fibonacci series) without running out of memory.
- Simplified Code: They provide an easy way to create iterators without the boilerplate code of implementing the __iter__() and __next__() methods manually.
- Pipelining: Generators can be chained together to form efficient data processing pipelines, where the output of one generator becomes the input for the next. 



````python
def count_up_to(limit):
    current = 1
    while current <= limit:
        yield current
        current += 1

# Using the generator in a for loop
# Each time the loop asks for the next value, the count_up_to function resumes, 
# executes the yield statement, pauses, and waits for the next request. 
number_generator = count_up_to(5)
print(type(number_generator))
for number in number_generator:
    print(number)
````

## Special Function Type: Async (asynchronous, coroutine, cooperative) Functions

In Python, an async function (or coroutine function) is a special type of function defined using the async def syntax, which allows for cooperative multitasking within a single thread. This enables the program to run other tasks while the function is waiting for I/O-bound operations (like network requests, file I/O, or database queries) to complete, thereby improving overall efficiency and responsiveness. 

Key Concepts
- async def: This keyword is used to declare an asynchronous function. Calling an async def function creates a coroutine object, but does not start its execution.
- await: This keyword can only be used inside an async def function. When encountered, it pauses the execution of the current coroutine and yields control back to the event loop, allowing it to run other pending tasks.
- Event Loop: This is the core of asynchronous programming in Python. It manages and distributes the execution of different tasks, switching between them when a task yields control with await. The standard library provides the asyncio module for this purpose.
- Coroutines: These are the "pausable" functions created by async def. They represent a computation that can be suspended and resumed later at specific points (marked by await).

How It Works  

Unlike synchronous programming, where tasks are executed sequentially and a function blocks the entire program while waiting for an operation, asynchronous functions utilize "cooperative multitasking". This means the function explicitly tells the event loop when it can take a break (using await) and let other work happen. 

For example, when an async function makes a network request:
1. It sends the request.
2. It hits an await statement (e.g., await session.get(url) using a library like aiohttp).
3. It yields control to the event loop.
4. The event loop then runs another task (e.g., handling a different request).
5. When the network response for the first task arrives, the event loop resumes that task. 

When to use Async Functions  

Async functions are best suited for I/O-bound operations (tasks that spend most of their time waiting for input/output), not CPU-bound tasks (tasks that use a lot of processing power). Common use cases include: 
- Network communication (HTTP requests, web servers)
- Database operations
- File reading/writing
- Managing a large number of concurrent connections 


````python
import asyncio

async def func():
    print("Start of function")
    await asyncio.sleep(1) # Simulate an I/O bound task (e.g., network call)
    print("End of function")

async def main():
    # Run the function using await within the main coroutine
    await func()

# This is how you start the event loop and run the top-level async function
#asyncio.run(main())
````

## Special Function Type: Function Decorator (or Decorating a Function)

Python function decorators are a powerful and elegant way to modify or extend the behavior of functions or methods without changing their actual code. A decorator is a function that takes another function as an argument, adds some functionality, and returns a new function (often called a "wrapper" function). 

<b>Using funtiontools.wraps</b>
For decorators to work with functions that accept arguments and return values, the inner wrapper function typically uses *args and **kwargs to accept any number of positional and keyword arguments, passing them to the original function. It is also a best practice to use @functools.wraps(func) within the decorator to preserve the original function's name and docstring, aiding in documentation and debugging. 

<b>Common Use Cases</b>
Decorators are widely used in application development to implement cross-cutting concerns in a clean, reusable way. 
- Logging: Tracking function calls, arguments, and return values for debugging.
- Timing: Measuring how long a function takes to execute, useful for performance analysis.
Authentication & Authorization: Restricting access to certain functions based on user permissions (commonly used in frameworks like Flask and Django).
- Caching/Memoization: Storing the results of expensive function calls and returning the cached result when the same inputs occur again (Python's built-in @functools.lru_cache is an example).
- Input Validation: Ensuring input arguments meet specific criteria before the function's core logic runs.
- Error Handling/Retry Logic: Automatically retrying failed network calls or handling exceptions gracefully.
- Frameworks: Used extensively in web frameworks for routing URLs to specific view functions, e.g., @app.route('/') in Flask. 

<b>Built-in Python Decorators</b>
Python also provides several built-in decorators for use within classes: 
- @staticmethod: Defines a method that does not receive the instance (self) or the class (cls) as an implicit first argument.
- @classmethod: Defines a method that receives the class itself (cls) as the implicit first argument.
- @property: Allows a method to be accessed like an attribute (getter), with optional @<property-name>.setter and @<property-name>.deleter methods for controlled attribute access. 
 

````python
# The original function
def say_whee():
    print("1 Whee!")

# Define the decorator
def my_decorator(func):
    def wrapper():
        print("Before function call")
        func()
        print("After function call")
    return wrapper

# Create a say_whee wrapper and call the decorated function
say_whee_wrapper = my_decorator(say_whee)
say_whee_wrapper()

# Same as the above  The @my_decorator syntax is "syntactic sugar" for the line say_whee_wrapper = my_decorator(say_whee)
@my_decorator
def say_whee_wrapper2():
    print("2 Whee!")
say_whee_wrapper2()
````

## Decorating a Function with Arguments

Decorating a function that takes arguments requires the inner wrapper function within the decorator to accept arbitrary positional (*args) and keyword (**kwargs) arguments, and pass them to the original function.

Key Components
- @functools.wraps(func): This is a best practice decorator to preserve the original function's name, docstring, and other metadata. Without it, the decorated function would appear to have the name wrapper.
- *args and **kwargs: These allow the wrapper function to accept any number of positional and keyword arguments, respectively. This makes the decorator reusable for any function signature.
- func(*args, **kwargs): The original function is called inside the wrapper, receiving all the arguments passed to the decorated function.
- return result: The wrapper must return the value returned by the decorated function; otherwise, the decorated function will implicitly return None. 


````python
import functools

def my_decorator(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        # Do something before the function call
        print(f"Calling function {func.__name__} with args: {args}, kwargs: {kwargs}")

        # Call the original function with its arguments and get the result
        result = func(*args, **kwargs)
        
        # Do something after the function call, potentially with the result
        print(f"Function {func.__name__} returned: {result}")

        # Return the result of the original function
        return result
    return wrapper

@my_decorator
def greet(name, greeting="Hello"):
    full_message = f"{greeting}, {name}!"
    return full_message

# Calling the decorated function
message = greet("World", greeting="Hi")
print(f"The final returned message is: {message}")
````

## Decorator with arguments

Creating a Python decorator with arguments requires an extra layer of function nesting to handle the arguments passed to the decorator itself. The outer function accepts the decorator's arguments and returns the actual decorator, which then works as a standard decorator, accepting the function to be decorated. 

The Three-Layer Structure   
The fundamental pattern for a decorator with arguments involves three nested functions: 
1. Outer Function (Decorator Factory): Takes the decorator-specific arguments (e.g., a number of times to repeat a function call). It returns the middle layer function.
2. Middle Function (The actual decorator): Takes the function being decorated (func) as an argument. It returns the inner wrapper function.
3. Inner Function (Wrapper): This function is the one that actually runs in place of the original function. It uses *args and **kwargs to accept any arguments passed to the decorated function and can access the decorator's arguments through closure.  

````python
import functools

def repeat(times):  # This is a Decorator Factory and it takes an argument
    """Decorator factory that takes arguments."""
    def decorator(func): # This is the Decorator
        """The actual decorator that takes the function."""
        @functools.wraps(func)
        def wrapper(*args, **kwargs): # This is the Wrapper
            """The wrapper function that runs the decorated function."""
            results = []
            for _ in range(times):
                results.append(func(*args, **kwargs))  # Call the wrapped function
            return results
        return wrapper
    return decorator

@repeat(times=3)
def say_hello(name):
    """Returns a greeting."""
    return f"Hello, {name}!"

# Call the decorated function
print(say_hello("World"))
````

## Argument Unpacking with * and ** operators

Python argument unpacking is a feature that uses the * and ** operators to break down iterable objects (such as lists, tuples, or dictionaries) so their elements can be passed as individual arguments to a function call. 


````python
# Unpacking Positional Arguments with *
def addition(a, b, c):
    return a + b + c

numbers_list = [1, 5, 10]
result = addition(*numbers_list)
print("Sum:", result)
````

````python
# Unpacking Keyword Arguments with **
def info(name, age, country):
    print(f"Name: {name}, Age: {age}, Country: {country}")

data_dict = {"name": "Alice", "age": 30, "country": "New York"}
info(**data_dict)    
````

````python
# Extended Unpacking: The asterisk syntax can also be used in assignment statements to capture
# multiple values into a single list variable, known as "extended unpacking" (available in Python 3).
first, *rest, last = [1, 2, 3, 4, 5]
print(first, rest, last)
````

````python
# Merging Iterables: The unpacking operators can be used to merge multiple iterables or dictionaries into a new one.
list_one = [1, 2]
list_two = [3, 4]
merged_list = [*list_one, *list_two] # merged_list is [1, 2, 3, 4]
print(merged_list)

dict_one = {'a': 1, 'b': 2}
dict_two = {'c': 3, 'd': 4}
merged_dict = {**dict_one, **dict_two} # merged_dict is {'a': 1, 'b': 2, 'c': 3, 'd': 4}
print(merged_dict)
````

## Generic Function

A Python "generic function" is a function that can operate on various data types while maintaining type consistency
and allowing static type checkers (like mypy) to verify type correctness. This is achieved using type hints 
and type variables from the typing module.

Defining a Generic Function

To create a generic function, you use a type variable as a placeholder for the actual type. 
The type checker infers the specific type when the function is called.

````python
# For Python 3.11 and older (old syntax):
from typing import TypeVar, Sequence

T = TypeVar("T")  # Define a type variable

def first(seq: Sequence[T]) -> T:
    """Returns the first element of a sequence."""
    return seq[0]

# Calling first([1, 2, 3]): The type checker infers T as int, and the return type is checked as int.
print(first([1,2,3]))

# Calling first(["a", "b"]): The type checker infers T as str, and the return type is checked as str. 
print(first(['a', 'b', 'c']))
````

````python
# Here is an example using the modern syntax (Python 3.12+):
def first[T](seq: Sequence[T]) -> T:
   """Returns the first element of a sequence."""
   return seq[0]

# Calling first([1, 2, 3]): The type checker infers T as int, and the return type is checked as int.
print(first([1,2,3]))

# Calling first(["a", "b"]): The type checker infers T as str, and the return type is checked as str. 
print(first(['a', 'b', 'c']))
````

## Generic

In Python, a generic function uses type hints with type variables (placeholders for types) to define a function that can operate on various data types while maintaining type consistency, which is checked by static analysis tools like mypy. 

````python
# For Python 3.11 and older (old syntax):
from typing import Generic, TypeVar, List

T = TypeVar("T")          # For Python 3.11 and older (old syntax)
                          # For Python 3.13 and later (not need to define T)

#class Stack[T]:          # For Python 3.12 and later (new syntax)
class Stack(Generic[T]):  # For Python 3.11 and older (old syntax)
    def __init__(self):
        self._items: List[T] = []

    def push(self, item: T) -> None:
        self._items.append(item)

    def pop(self) -> T:
        if not self._items:
            raise IndexError("pop from empty stack")
        return self._items.pop()

    def peek(self) -> T:
        if not self._items:
            raise IndexError("peek from empty stack")
        return self._items[-1]

# Create a Stack of integer
int_stack: Stack[int] = Stack()
int_stack.push(10)
int_stack.push(20)
print(int_stack.pop())

# Create a Stack of string
str_stack: Stack[str] = Stack()
str_stack.push("a")
str_stack.push("b")
print(str_stack.pop())
````

Single leading underscore|	_variable|	Indicates that the name is meant for internal use only|
Single trailing underscore|	class_|	Avoids naming conflicts with Python keywords and built-in names|
Double leading underscore|	__attribute|	Triggers name mangling in the context of Python classes|
Double leading and trailing underscore|	__ name __|	Indicates special attributes and methods that Python provides|
Single underscore|	_	|Indicates a temporary or throwaway variable|

````python
from annotated_doc import Doc

from typing import (
    Annotated,
    Any,
    TypeVar,
)

from annotated_doc import Doc

ApiType = TypeVar("ApiType", bound="MyAPI")

class MyAPI:

    ############################################################
    # Constructor
    ############################################################
    def __init__(
        self: ApiType,
        debug: Annotated[
            bool,
            Doc(
                """
                Boolean indicating if debug logging should be returned on.  Default is false.
                """
            ),
        ] = False,
        name: Annotated[
            Any,
            Doc(
                """
                Setting the default name.  Default is 'Friend'.
                """
            ),
        ] = "Friend"
    ) -> None:
        """
            Function description
        """
        self.debug = debug
        self.name = name
        
    ############################################################
    # Method
    ############################################################
    def say_hi(
        self: ApiType,
        name: Annotated[
            Any,
            Doc(
                """
                Name of a person.  Default is set at the constructor.
                """
            ),
        ] = None
    ) -> None:
        """
            Function description
        """
        if self.debug:
            print(f"DEBUG: say_hi is called")
        if name is None:
            name = self.name
        print(f"Hi, {name}")

# Examples of using the API
api=MyAPI(debug=True, name="My Friend")
api.say_hi("Trung")
api.say_hi(1)
api.say_hi()
````

````python
# Given 2 lists
l1 = [299, 99, 399, 199]
l2 = [400, 100, 200, 300]

# Merge them to a single list, then sort the elements
l = sorted(l1 + l2)  # method 1: ADDing
print(l)

l1.extend(l2)        # method 2: Extend one of the list with values from the other list
print(sorted(l1))
````

````python
# Given 2 lists
l1 = ["a", "b", "c", "d"]
l2 = [100, 200, 300, 400]

# Convert them into dictionary
# See my other module: Python Tranforming iterables
d1 = dict(zip(l1, l2))  # zip to pair (or make tuples), then use dict (fuction or constructor) to make a dictionary
d2 = dict(zip(l2, l1))  # zip to pair (or make tuples), then use dict (fuction or constructor) to make a dictionary
print(d1)
print(d2)

# 3 ways to merge dictionaries
print(d1|d2)        # method 1: ORing
print({**d1, **d2}) # method 2: unpacking the dictionaries then put all pairs back to a dictionary
d1.update(d2)       # method 3: update. If key exists in both dictidonaries, the value from the rightmost dictionary is kept
print(d1)
````

````python
# Given 2 lists
l1 = ["a", "b", "c", "d"]
l2 = [100, 200, 300, 400]

# Convert them to list of tuples
# See my other module: Python Tranforming iterables
l = list(zip(l1, l2))  # zip to pair (or make tuples), then use list function (or constructor) to make a list
print(l)
````

````python
# Given a list of numbers:
nums = [1, 8, 5, 9, 7, 1, 2]

# select even numbers that are less than 10, then double each element and sort them
evens = sorted([x*2 for x in nums if x %2 == 0 and x < 10]) # use 'or' for ORing conditions
print(evens)
````
