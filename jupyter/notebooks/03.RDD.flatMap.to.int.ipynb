```python
from pyspark import SparkContext
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SparkSQL").getOrCreate()
sc = spark.sparkContext

data = sc.parallelize([1, 4, 7])
print ()
print ("INPUT:")
for result in data.collect():
    print (result)

print ()
print ("map (transform) each element to a list: lambda x: [x,x+1,x+2]")
results = data.map(lambda x: [x,x+1,x+2]).collect()  # map a scalar x to a list
for result in results:
    print (result)
print ("Note that the result above, each row (the result of the lamba function) is a list")

print
print ("flatMap (transform) each element to multi elements: lambda x: [x,x+1,x+2]")
results = data.flatMap(lambda x: [x,x+1,x+2]).collect()  # map a scalar x to a list, then flattern the list
for result in results:
    print (result)
print ("Note that the result above, each row is an integer")
```
