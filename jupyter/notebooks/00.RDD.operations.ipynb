## Resilient Distributed Dataset (RDD) operations

Resilient Distributed Dataset (RDD) operations in Apache Spark consist of
- Transformations (creating new RDDs lazily) and
- Actions (triggering computation and returning results).
  
Because RDDs are immutable, transformations never change the original RDD, but instead create a new one. These operations are executed in parallel across a cluster.    

Reference: https://spark.apache.org/docs/latest/rdd-programming-guide.html


### Transformations (Lazy Evaluation)

These operations define a new RDD from an existing one without immediate execution.
- map(func): Returns a new RDD by applying a function to each element.
- filter(func): Returns a new RDD containing only elements that pass a predicate function.
- flatMap(func): Similar to map, but each input item can be mapped to 0 or more output items.
- mapPartitions(func): Maps each partition (a more efficient version of map).
- distinct(): Returns a new RDD with unique elements.
- reduceByKey(func): (Pair RDDs) Merges values for each key using an associative reduce function.
- join(other): (Pair RDDs) Performs an inner join.


### Actions (Result Generation)

These operations trigger the computation of transformations and return results to the driver or write to storage.
- collect(): Returns all elements of the dataset to the driver program.
- reduce(func): Aggregates elements using a function.
- count(): Returns the number of elements in the RDD.
- saveAsTextFile(path): Writes elements to a file in HDFS or local file system.
- take(n): Returns the first \(n\) elements.


### Key Characteristics

- Immutability: RDDs cannot be changed once created.
- Lineage: Spark tracks the "lineage" of transformations to recover data automatically if a partition is lost.
- Lazy Evaluation: Transformations are not executed until an action is called, allowing Spark to optimize the execution plan.

## Pair RDD operations

Pair RDD operations in Apache Spark are specialized transformations and actions for RDDs containing key-value pairs, enabling parallel, key-based processing and data shuffling. Key operations include 
- reduceByKey (aggregate),
- groupByKey (group),
- sortByKey (sort), and
- join (merge), allowing efficient manipulation of data across the network.


### Core Pair RDD Transformations

- reduceByKey(func): Merges values for each key using an associative and commutative reduce function.
- groupByKey(): Groups values for each key into a single sequence.
- sortByKey(): Sorts the RDD by its keys.
- join(otherRDD): Performs an inner join between two pair RDDs.
- mapValues(func): Applies a function to each value in the pair RDD without changing the key.
- flatMapValues(func): Maps each value to a sequence of values, creating a new pair RDD with the same keys.
- combineByKey(createCombiner, mergeValue, mergeCombiners): A versatile function for combining elements based on keys.

### Key Pair RDD Actions

- countByKey(): Counts the number of elements for each key.
- collectAsMap(): Returns the key-value pairs as a map to the driver program.
- lookup(key): Returns all values associated with the provided key.

### Key Characteristics

- Shuffling: Many pair RDD operations, such as reduceByKey and join, trigger a "shuffle," which involves ***transferring data across the network***.
- Tuple Representation: In Scala, these operations are available on RDDs of Tuple2 objects.
- Efficiency: ***reduceByKey is generally preferred*** over groupByKey because it performs local reductions before shuffling data.
