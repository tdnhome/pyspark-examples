## Spark Actions

In Apache Spark, actions are operations that trigger the actual computation of data, returning a result to the driver or writing data to external storage. While both RDD (Resilient Distributed Dataset) and DataFrames use actions similarly to trigger lazy evaluation, they differ significantly in optimization and structure.

Examples: https://medium.com/@kiranvutukuri/spark-actions-part-12-60fc063efe81


### Key Differences at a Glance

|Feature |RDD Actions |DataFrame Actions |
|------- |----------- |----------------- |
|Optimization |None. Executes exactly as written. | Uses Catalyst Optimizer to create a highly efficient execution plan.|
|Data Format  |Unstructured; treats data as raw JVM objects. |Structured; organizes data into named columns like a database table.|
|Level of Control |Low-level. Fine-grained control over individual objects. | High-level. Declarative API focused on "what" to do rather than "how".|
| Performance  |Slower for structured tasks due to object serialization. |Faster due to off-heap memory and binary format (Tungsten).|

### Common Actions Comparison
Both APIs share many similar actions, but they are implemented through different APIs.
1. ***Data Retrieval Actions***
- RDD: collect(), take(n), first(), top(n).
- DataFrame: collect(), take(n), first(), head(), show().
- Note: DataFrame's show() is a unique action that prints a formatted table to the console.
2. ***Aggregation Actions***
- RDD: count(), reduce(), fold(), aggregate().
- DataFrame: count(), describe(), summary().
- Note: Most complex aggregations in DataFrames are actually transformations (like groupBy().agg()) that are only computed when a final action is called.
3. ***Saving Data***
- RDD: saveAsTextFile(), saveAsObjectFile(), saveAsSequenceFile().
- DataFrame: write.parquet(), write.csv(), write.json(), write.saveAsTable().

### When to Use Each
- Use DataFrame Actions for most ***real-world*** engineering tasks. They are significantly faster because Spark's optimizer can reorder filters, prune columns, and manage memory more efficiently.
- Use RDD Actions only if you need ***low-level control over unstructured data*** (like media streams), need to implement custom partitioning, or are maintaining legacy code.
- You can convert between the two at any time using ***rdd.toDF()*** to gain optimization or ***df.rdd*** to regain low-level control.
