```python
# https://medium.com/@rganesh0203/select-vs-selectexpr-in-pyspark-46eeed92d016

from pyspark.sql import SparkSession

# Create SparkSession
print("https://medium.com/@rganesh0203/select-vs-selectexpr-in-pyspark-46eeed92d016")

spark = SparkSession.builder.appName("SparkByExamples").getOrCreate()

# Create DataFrame
data = [
    ("Alice", 30, 5000),
    ("Bob", 45, 8000),
    ("Charlie", 35, 6000)
]

columns = ["Name", "Age", "Salary"]

df = spark.createDataFrame(data, columns)
df.show()
```

select():
1. Purpose: Used to select specific columns from a DataFrame by directly referencing their names.
2. Usage: This method is straightforward and is used when you just need to select columns or apply simple transformations like renaming.
3. Key Point: select() is useful when you need to perform column selection or basic transformations.

```python
from pyspark.sql.functions import col

# Selecting the Name and Age columns
df.select("Name", "Age").show()

# Renaming the "Name" column to "EmployeeName"
df.select(col("Name").alias("EmployeeName"), "Age").show()
```

selectExpr():
1. Purpose: Allows you to select columns using SQL expressions. It is more powerful as it lets you perform transformations and calculations directly within the method.
2. Usage: This method is particularly useful when you need to apply SQL-like expressions on columns (e.g., performing aggregations, transformations, or using SQL functions).
3. Key Point: selectExpr() is powerful for complex transformations, calculations, or applying SQL functions inline, allowing more flexibility in how data is selected and manipulated.

```python
# Selecting the Name column and doubling the Salary
df.selectExpr("Name", "Salary * 2 as DoubleSalary").show()

# Applying SQL functions: calculating the average of Age and summing Salary
df.selectExpr("avg(Age) as Average_Age", "sum(Salary) as Total_Salary").show()
```
