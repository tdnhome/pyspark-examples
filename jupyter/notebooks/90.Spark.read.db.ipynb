```python
import os
# Replace '8.0.33' with your desired version
os.environ['PYSPARK_SUBMIT_ARGS'] = '--packages mysql:mysql-connector-java:8.0.33 pyspark-shell'
```

```python
from pyspark.sql import SparkSession

# Create SparkSession
print("Spark.SQL.Example: https://sparkbyexamples.com/pyspark/pyspark-sql-with-examples")
spark = SparkSession.builder.appName("SparkByExamples").getOrCreate()

# Connection info
jdbc_url = "jdbc:mysql://localhost:3306/db-demo"
table_name = "customer"
connection_properties = {
    "user": "demouser",
    "password": "demopass",
    "driver": "com.mysql.cj.jdbc.Driver" # Use "com.mysql.jdbc.Driver" for older versions of the connector
}

# Read from JDBC 
df = spark.read.jdbc(
    url=jdbc_url,
    table=table_name,
    properties=connection_properties
)


print("DF.printSchema")
df.printSchema()

print("DF.show")
df.show()
```
