```python
# Groups all values for each key into a single iterable collection (a list). 

from pyspark import SparkContext
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SparkSQL").getOrCreate()
sc = spark.sparkContext

data = [("fruit", "apple"), ("fruit", "banana"), ("veg", "carrot"), ("fruit", "orange")]
rdd = sc.parallelize(data)

# Group the values by key
grouped_rdd = rdd.groupByKey()

# Convert the iterables to lists for display
result = grouped_rdd.mapValues(list).collect()

print(result)
# Output: [('veg', ['carrot']), ('fruit', ['apple', 'banana', 'orange'])]
```
