### Read parquet file to panda DataFrame

```python
import pandas as pd
df = pd.read_parquet('person.parquet')
print(df)
```

### Read parquet file to Spark DataFrame

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
df = spark.read.parquet('person.parquet')
df.printSchema()
df.show()
```
