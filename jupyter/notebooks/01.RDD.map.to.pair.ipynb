```python
from pyspark import SparkContext
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SparkSQL").getOrCreate()
sc = spark.sparkContext

# This funtion naps data into key/value pairs (age, numFriends)
def parseLine(line):
    fields = line.split(',')
    age = int(fields[2])
    numFriends = int(fields[3])
    return (age, numFriends)

# Read all data into collection of line
lines = sc.textFile("fakefriends_small.csv")

# Map data into key/value pairs (age, numFriends)
# java: JavaPairRDD<Integer, Integer> ageToNumFriends = lines.mapToPair(...);
print ('Making a map tuple (age,numFriends) which is a set of key/value pairs')
ageToNumFriends = lines.map(parseLine)

results = ageToNumFriends.collect()
for result in results:
    print (result)
```
