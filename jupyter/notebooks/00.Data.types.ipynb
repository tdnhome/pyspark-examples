## Built-in Data Types

In programming, data type is an important concept.

Variables can store data of different types, and different types can do different things.

Python has the following data types built-in by default, in these categories:

- Text Type:	    str
- Numeric Types:	int, float, complex
- Sequence Types:	list, tuple, range
- Mapping Type:	    dict
- Set Types:	    set, frozenset
- Boolean Type:	    bool
- Binary Types:	    bytes, bytearray, memoryview
- None Type:	    NoneType

```python
# In Python, the data type is set when you assign a value to a variable:
x = "Hello World"	# str	
x = 20	            # int	
x = 20.5	        # float	
x = 1j	            # complex	
x = ["apple", "banana", "cherry"]	# list using square brackets []	
x = ("apple", "banana", "cherry")	# tuple using parentheses ()	
x = range(6)	                    # range using range() function	
x = {"name" : "John", "age" : 36}	# dict using brackets {k:v, k:v}	
x = {"apple", "banana", "cherry"}	# set using brackets {v,v,v}	
x = frozenset({"apple", "banana", "cherry"})	# frozenset
x = True	                        # bool	
x = b"Hello"	                    # bytes
x = bytearray(5)	                # bytearray	
x = memoryview(bytes(5))	        # memoryview	
x = None	                        # NoneType	
```

```python
# Setting the Specific Data Type
# If you want to specify the data type, you can use the following constructor functions:
x = str("Hello World")		
x = int(20)
x = float(20.5)	
x = complex(1j)	
x = list(("apple", "banana", "cherry"))
x = tuple(("apple", "banana", "cherry"))	
x = range(6)
x = dict(name="John", age=36)
x = set(("apple", "banana", "cherry"))
x = frozenset(("apple", "banana", "cherry"))
x = bool(5)
x = bytes(5)	
x = bytearray(5)	
x = memoryview(bytes(5))
```

## Binary Data Type

In Python, “binary data types” refer to types designed to store and manipulate raw bytes (binary data) 
rather than human-readable text. These are especially useful for file I/O, networking, serialization, 
and working with low-level data.

The core binary data types in Python are:

|Type	|Use Case|
|-------|--------|
|bytes	|Read-only binary data (files, network packets)|
|bytearray	|Modifiable binary buffers|
|memoryview	|High-performance slicing / no-copy operations|

### bytes

Immutable sequence of bytes (0–255)

Think of bytes as a read-only array of integers where each value represents a byte.

Key properties
- Immutable (cannot be changed after creation)
- Each element is an integer in range 0–255
- Often used for binary files, network packets, encoded text

```python
# Creating bytes
b1 = b"hello"             # literal
b2 = bytes([65, 66, 67])  # from integers → b'ABC'
b3 = bytes(5)             # 5 zero bytes → b'\x00\x00\x00\x00\x00'

# Accessing data
b = b"ABC"
print(b[0])   # 65 (ASCII of 'A')
print(b[0:2]) # b'AB'

# Common methods
print(b.lower())      # lower case
print(b.find(b"A"))   # return index of A
print(b.replace(b"A", b"Z"))
print(b) 
```

### bytearray

Mutable version of bytes.  This is like bytes, but you can modify it in-place.

Key properties
- Mutable
- Same value range: 0–255
- Useful when you need to modify binary data efficiently

Why use it?
- Efficient for building or editing binary streams
- Avoids creating new objects (unlike bytes)

```python
# Creating bytearray
ba = bytearray(b"hello")

# Modifying data
ba[0] = 72        # change 'h' → 'H'
ba.append(33)     # add '!'
print(ba)         # bytearray(b'Hello!')
```

### memoryview

Zero-copy view of binary data

A memoryview lets you access the internal memory of another object without copying it.

Key properties
- No data copy → high performance
- Works with bytes, bytearray, and other buffer-supporting objects
- Useful for large datasets or performance-critical code

```python
data = bytearray(b"abcdef")
mv = memoryview(data)

print(mv[0])     # 97 ('a')
mv[0] = 65       # modify original data via view

print(data)      # bytearray(b'Abcdef')
```

### Encoding vs Binary Data

A common confusion: text vs binary
- str → text (Unicode)
- bytes → binary (raw data)

```python
# Convert between them
text = "hello"
print(text)

# encode (str → bytes)
b = text.encode("utf-8")
print(b)

# decode (bytes → str)
text2 = b.decode("utf-8")
print(text2)
```
