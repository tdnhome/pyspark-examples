### Introduction to Python Lambda 


A lambda function in Python is a small, anonymous function defined without a name. Unlike standard functions defined with def, lambdas are restricted to a single expression and are typically used for short-term, throwaway tasks.

Key Characteristics

- Nameless: You don't have to name the function if you only need it once.
- One Expression Only: They cannot contain multiple lines or statements (like return or loops).
- Implicit Return: The result of the expression is automatically returned.

Syntax: lambda arguments expression condition
- agruments: can be 0 or more
- expression: any valid python expression
- condition: if else
  - Example below says: return a if a<b else return b
  - lambda a, b: a if a<b else b

Limitations

- No multiple expressions
- No statements like if, for, while, try, =
- However, they can include conditional expressions:
  - max_val = lambda a, b: a if a > b else b
- Harder to read if overused

### Basic examples

```python
# Normal function using def syntax
def square_func(x):
    return x*x
print(square_func(5))

# Function using lambda syntax
square_lambda = lambda x: x * x
print(square_lambda(5))
```

### Multiple parameters and condiction

```python
# Lambda with multiple agrument and condition
smaller = lambda x,y: x if x<y else y # return x if x<y; else return y
print(smaller(5, 6))

# Another example of condition on lambda 
check_even = lambda x: "Even" if x % 2 == 0 else "Odd"
print(check_even(10))

# No agrument
say_hi = lambda: "Hi"
print(say_hi())
```

### Common Use Cases

Lambdas are most powerful when used as arguments in "higher-order" functions like 
- map()
- filter()
- sorted()
- functools.reduce()

```python
# Using with map() (transform data) - applies the function to every item in a list.
nums = [1, 2, 3, 4]
doubled = list(map(lambda x: x * 2, nums))
print(doubled)
```

```python
# Using with filter() (select data) - keeps only the items for which the function returns True.
nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)
```

```python
# Using with sorted() (custom sorting) - customizes the sorting criteria, such as sorting a list of tuples by their second element.
students = [("Alice", 25), ("Bob", 20), ("Charlie", 23)]

print(f"default sorting: {sorted(students)}") # default sorting: by first element

sorted_by_age = sorted(students, key=lambda student: student[1])
print(f"custom sorting:  {sorted_by_age}") 
```

```python
# With functools.reduce() - aggregate values on the list
from functools import reduce
nums = [1, 2, 3, 4]
sum_all = reduce(lambda x, y: x + y, nums)
print(sum_all)
```
