```python
# Google: pyspark map to pairrdd example

from pyspark import SparkContext
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SparkSQL").getOrCreate()
sc = spark.sparkContext


# 1. Create a regular RDD from a list of strings
data = ["pyspark rdd example", "pyspark pair rdd", "pyspark map to pairrdd"]
rdd = sc.parallelize(data)
print("\nData: elements are sentences (list of words):")
print(rdd.collect())

# 2. Use flatMap to split each string into words and create a new RDD of single words
words_rdd = rdd.flatMap(lambda line: line.split(" "))
print("\nRDD.flatMap sentences to words")
print(words_rdd.collect())
# would output: ['pyspark', 'rdd', 'example', 'pyspark', 'pair', 'rdd', 'pyspark', 'map', 'to', 'pairrdd']

# 3. Use map to transform each word into a (key, value) tuple, where the word is the key and 1 is the value
# This creates the Pair RDD
pair_rdd = words_rdd.map(lambda word: (word, 1))
print("\nRDD.map to pair: (word, 1)")
print(pair_rdd.collect())
# would output: [('pyspark', 1), ('rdd', 1), ('example', 1), ('pyspark', 1), ('pair', 1), ('rdd', 1), ('pyspark', 1), ('map', 1), ('to', 1), ('pairrdd', 1)]

# 4. (Optional) Use a Pair RDD operation like reduceByKey to count occurrences of each word
word_counts = pair_rdd.reduceByKey(lambda a, b: a + b)

# 5. Collect and print the results
print("\nWord counts using reduceByKey on the Pair RDD:")
print(word_counts.collect())
```
