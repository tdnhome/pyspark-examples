```python
from pyspark.sql import SparkSession

# Create SparkSession
print("Spark.SQL.Example: https://sparkbyexamples.com/pyspark/pyspark-sql-with-examples")

spark = SparkSession.builder.appName("SparkByExamples").getOrCreate()

# Create DataFrame
# Important: .option must come before .csv
# Line Continuation: 
#   A backslash at the end of a line tells the Python interpreter to ignore the 
#   newline and continue the statement on the next line.  There should be no space after lthe backslash
print("spark.read.csv")
df = spark.read \
    .option("header",True) \
    .csv("simple-zipcodes.csv")

print("DF.printSchema")
df.printSchema()

print("DF.show")
df.show()

print("DF.write.json('simple-zipcodes-from-df.json')")
df.write \
    .mode("overwrite") \
    .json("simple-zipcodes-from-df.json")
print("DF.write.json: SUCCESSFUL")
```
