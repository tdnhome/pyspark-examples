```python
#----------------------------------------------------------------------
# To run: 
#    1. Open command prompt
#    2. cd to this folder
#    3. spark-submit python_script.py
#----------------------------------------------------------------------
#
# http://spark.apache.org/docs/latest/programming-guide.html
#
# map(func)
#
# Return a new distributed dataset formed by passing 
# each element of the source through a function
#  function(x) returning y
# Each input item is mapped to 1 output item. 
#
#----------------------------------------------------------------------
print()
print("===================================================")
print("RDD.map(funcReturningString)")
print("---------------------------------------------------")
print("Return a new distributed dataset formed by passing")
print("each element of the source through a function func.")
print("Each input item is mapped to a string")
print("Example: rddList = rdd.map(lambda x: x.upper())")
print("===================================================")

from pyspark import SparkContext
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SparkSQL").getOrCreate()
sc = spark.sparkContext

# Read input
lines = sc.textFile("text.txt")

# Print the input
print()
print("INPUT:")
results = lines.collect()
for result in results:
    print(result)

# Parse data
parsedLines = lines.map(lambda x: x.upper()) # map string to string

# Collect and print the results
print()
print("MAPPED:")
results = parsedLines.collect()
for result in results:
    print(result)
```
