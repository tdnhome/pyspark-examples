```python
# https://sparkbyexamples.com/pyspark/pyspark-string-functions-with-examples/

from pyspark.sql import SparkSession
from pyspark.sql.functions import *

# Create SparkSession
spark = SparkSession.builder \
            .appName('SparkByExamples.com') \
            .getOrCreate()

data = [('James','','Smith','1991-04-01','M',3000),
  ('Michael','Rose','','2000-05-19','M',4000),
  ('Robert','','Williams','1978-09-05','M',4000),
  ('Maria','Anne','Jones','1967-12-01','F',4000),
  ('Jen','Mary','Brown','1980-02-17','F',-1)
]
columns = ["firstname","middlename","lastname","dob","gender","salary"]
df = spark.createDataFrame(data=data, schema = columns)
print("df.show:")
df.show(truncate=False)

# Using select() with concat()
print("concat:")
df2=df.select(concat(df.firstname,df.middlename,df.lastname)
              .alias("FullName"),"dob","gender","salary")
df2.show(truncate=False)

# concat_ws(): similar to concat but with a specified separator into a single string column
print("concat_ws:")
separator=' '
df3=df.select(concat_ws(separator,df.firstname,df.middlename,df.lastname)
              .alias("FullName"),"dob","gender","salary")
df3.show(truncate=False)

# substr(), substring(): get the substring of a column
print("substr, substring:")
data = [("John",), ("Jane",), ("Robert",)]
columns = ["name"]
df = spark.createDataFrame(data, columns)

# Add a new column with the substr and substring functions
df_with_substr = df \
    .withColumn("substr", expr("substr(name, 2, 3)")) \
    .withColumn("substring", expr("substring(name, 2, 3)"))

df_with_substr.show()

# split()
print("split:")
data = [("John,Doe",), ("Jane,Smith",), ("Robert,Johnson",)]
columns = ["full_name"]

df = spark.createDataFrame(data, columns)

# Use the split function to split the "full_name" column by comma
split_columns = split(df["full_name"], ",")

# Add the split columns to the DataFrame
df_with_split = df \
    .withColumn("first_name", split_columns[0]) \
    .withColumn("last_name", split_columns[1])

df_with_split.show()

# contains:
print("filter with contains:")
data = [("John Doe",), ("Jane Smith",), ("Robert Johnson",)]
columns = ["full_name"]

df = spark.createDataFrame(data, columns)

# Use filter and contains to check if the column contains the specified substring
substring_to_check = "Smith"
filtered_df = df.filter(col("full_name").contains(substring_to_check))
filtered_df.show()

# regexp_replace: to replace spaces with underscores
print("regexp_replace: space with _")
df_replaced = df.withColumn("name_with_underscore", regexp_replace(df["full_name"], r' ', '_'))

# Show the DataFrame
df_replaced.show()

```
