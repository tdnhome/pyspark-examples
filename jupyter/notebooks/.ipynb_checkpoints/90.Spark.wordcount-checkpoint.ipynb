## Example of word count using Spark DataFrame

### Version 1 - Better explanation

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import explode, split, col

# 1. Initialize SparkSession
spark = SparkSession.builder.appName("WordCount").getOrCreate()

# 2. Load data into a DataFrame
df = spark.read.text("Book.txt")

# 3. Select the "value" column (lines), split them into words and explode them into separate rows
# - col(): function to reference a specific column in a DataFrame by its name
# - split(): Breaks each line into an array of words
# - explode(): Flattens the array so each word gets its own row
# By default, spark.read.text reads each line into a column named 'value'
words_df = df.select(explode(split(col("value"), " ")).alias("word"))

# 4. Group by the word column and count occurrences
word_counts = words_df.groupBy("word").count()

# 5. Show the results (sorted by count descending)
word_counts.orderBy(col("count").desc()).show()
```

### Version 2 - More compact

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import explode, split, col

# 1. Initialize Spark Session
spark = SparkSession.builder.appName("WordCountExample").getOrCreate()

# 2. Read text file into a DataFrame
df = spark.read.text("Book.txt")

# 3. Transform and Count
# - col(): function to reference a specific column in a DataFrame by its name
# - split(): Breaks each line into an array of words
# - explode(): Flattens the array so each word gets its own row
# - df.withColumn(): create a new column on the exiting DataFrame
# By default, spark.read.text reads each line into a column named 'value'
word_counts = df.withColumn("word", explode(split(col("value"), " "))) \
    .groupBy("word") \
    .count() \
    .orderBy(col("count").desc())

# 4. Display results
word_counts.show()
```
