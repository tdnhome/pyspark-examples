## JSON File Handling

Reference: https://www.geeksforgeeks.org/python/json-load-in-python/

1. json.load() is a function from Python's built-in json module. It reads JSON data from a file and converts it into the corresponding Python object.

- JSON objects: dict
- JSON arrays: list
- JSON strings, numbers, booleans: Corresponding Python types

2. json.loads() is a function to read data from a string


### Read a single json file

```python
# json.load(): Load file with single json - Open the file and load the data
import json
#with open('simple-zipcode.json', 'r', encoding='utf-8') as file:
with open('simple-zipcode.json') as file:
    data = json.load(file)
print(data)
```

### Read multiple jsons in a file, each line is a valid json file

```python
# json.loads(): Load from string having single json 
import json
json_list = []

# Parse each line of the file into json list.  Each row is a dictionary (returned from json.loads function)
with open('simple-zipcodes.json') as file:
    for line in file:
        # Load each individual line as a dictionary
        json_list.append(json.loads(line))

# Print the json list 
for json in json_list:
    print()
    for key, value in json.items():
        print(f"{key}: {value}")
```
