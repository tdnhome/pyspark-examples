### Syntax
Python uses two primary types of loops to automate repetitive tasks. 
Both rely on indentation to define the block of code to be repeated. 

- for loops
- while loops



### for loop

```python
# Loop over a range
for i in range(3):  # Iterates from 0 to 2
    print(i)
```

```python
# Loop over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
# Loop over a list with break
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print(x)
    if x == "banana":
        break
```

```python
# Loop over a list with continue (go to next loop)
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    if x == "banana":
        continue
    print(x)
```

```python
# The else keyword in a for loop specifies a block of code to be executed when the loop is finished
for x in range(3):
  print(x)
else:
  print("Finally finished!")
    
```

```python
# For loop with break and else - The else is not executed if the loop is broken (break statement executed)
for x in range(3):
  if x == 2: break
  print(x)
else:
  print("Finally finished!")
```

```python
# for loops cannot be empty, but if you for some reason have a for loop with no content, 
# put in the pass statement to avoid getting an error.
for x in [0, 1, 2]:
    pass
```

```python
# Loop over a dictionary
my_dict = {"a": 1, "b": 2, "c": 3}
print("Keys:")
for key in my_dict:
    print(key)
    
print("Values:")
for value in my_dict.values():
    print(value)

print("Keys and values:")
for key, value in my_dict.items():
    print(f"Key: {key}, Value: {value}")
```

### while loop

- A while loop will continue until the statement is false.
- A while loop will go into infinity loop if the condition does not become false 

```python
x = 0
while x < 3:
  print(x)
  x = x + 1
```

```python
with open("text.txt", "r") as file:
    content = file.read()
    print(content)
```
