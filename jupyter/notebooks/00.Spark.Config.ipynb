## Spark Config

Spark provides three locations to configure the system. If the same property is defined in multiple places, Spark follows this hierarchy (highest to lowest):

- Properties set directly on the SparkConf object in your code or through Java system properties.
- Flags passed to the spark-submit command
- spark-submit command also take properties file specified by --properties-file overriding conf/spark-defaults.conf.
- Logging can be configured through log4j2.properties.

References:

```python
from pyspark import SparkContext
from pyspark.conf import SparkConf


# 1. Create a SparkConf object and set configurations
conf = SparkConf() \
    .setAppName("PySpark App") \
    .setMaster("local[2]") \
    .set("spark.executor.memory", "1g") \
    .set("spark.executor.core", "2") 


# 2. Initialize SparkContext with the configuration object: RDD uses SparkContext
sc = SparkContext(conf=conf)

# Check values from the configuration
print(f"App Name: {sc.appName}")
print(f"Master: {sc.master}")

# Close the context and shut down the application
sc.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark import SparkConf

# 1. Create a SparkConf object and set configurations
conf = SparkConf() \
    .setMaster("local[2]") \
    .set("spark.executor.memory", "1g") \
    .set("spark.executor.core", "2") 

# 2. Initialize SparkContext with the configuration object: RDD uses SparkContext

# Pass the SparkConf object to the builder
spark = SparkSession.builder \
    .config(conf=conf) \
    .getOrCreate()

# Check values from the configuration
print(f"spark.executor.memory: ${spark.conf.get('spark.executor.memory')}")
print(f"spark.executor.core: ${spark.conf.get('spark.executor.core')}")

# Close the session and shut down the application
spark.stop()
```
