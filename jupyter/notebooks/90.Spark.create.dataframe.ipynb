## Create DataFrame from a list of tuples and list of columns matching the tuples' size

```python
from pyspark.sql import SparkSession

# Create SparkSession
spark = SparkSession.builder.appName("SparkByExamples").getOrCreate()

# From a Python List or Tuple.
# This is the most direct way to create a DataFrame from in-memory data. 
# You can optionally provide column names.
data = [("James", 34, "M"), ("Maria", 25, "F"), ("Jen", 31, "F")]
columns = ["Name", "Age", "Gender"]
df = spark.createDataFrame(data, schema=columns)
df.show()
```

## Create DataFrame from a list of dictionaries: keys => column name, values => column data 

```python
# https://spark.apache.org/docs/latest/api/python/reference/pyspark.sql/api/pyspark.sql.SparkSession.createDataFrame.html 
# From a dictionaly
d = [{'name': 'Alice', 'age': 19}, {'name': 'Jane', 'age': 21}]
spark.createDataFrame(d).show()
```

## Create DataFrame from a RDD using:
1. rdd.toDF(column_name_list)
2. sparkSession.createDataFrame(rdd,column_name_list)

```python
# From an RDD (Resilient Distributed Dataset)
# If you are working with the lower-level RDD API, you can convert an RDD to 
# a DataFrame using the .toDF() method or spark.createDataFrame().

rdd = spark.sparkContext.parallelize([("Raj", 1800), ("Priya", 2000)])
columns = ["Name", "Salary"]
df = rdd.toDF(columns)
df.show()

df = spark.createDataFrame(rdd, columns)
df.show()
```

## Create DataFrame using spark.range()
1. rdd.toDF(column_name_list)
2. sparkSession.createDataFrame(rdd,column_name_list)

```python
# Using spark.range()
# To create a simple DataFrame with a single column of sequential integer values 
# (useful for testing or synthetic data).
df = spark.range(1, 10, 2) # start, end (exclusive), step
df.show()
```

```python
# From a list of Row objects 
from pyspark.sql import Row
Person = Row('name', 'age')
df = spark.createDataFrame([Person("Alice", 1), Person("John", 2)])
df.show()
```
