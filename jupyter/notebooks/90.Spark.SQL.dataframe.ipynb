## Basic: Read from CSV file, show schema, show data

```python
# https://sparkbyexamples.com/pyspark/pyspark-sql-with-examples

from pyspark.sql import SparkSession

print("pyspark Dataframe API: https://spark.apache.org/docs/latest/api/python/reference/pyspark.sql/dataframe.html")
print("pyspark Tutorial: https://sparkbyexamples.com/pyspark-tutorial")
print("pyspark SQL Example: https://sparkbyexamples.com/pyspark/pyspark-sql-with-examples")

# Create SparkSession
spark = SparkSession.builder.appName("SparkByExamples").getOrCreate()

# Create DataFrame
print("spark.read.csv")
df = spark.read \
          .option("header",True) \
          .csv("simple-zipcodes.csv")

print("DF.printSchema")
df.printSchema()

print("DF.show")
df.show()
```

## DF.select vs spark.sql('SELECT')

```python
# DF.select
print("DF.select")
df.select("country","city","zipcode","state") \
  .show(5)
```

```python
# Create SQL table for spark.sql select statements
df.createOrReplaceTempView("Zipcodes")

# spark.sql.select
print("spark.sql(select)")
spark.sql("SELECT  country, city, zipcode, state FROM ZIPCODES") \
     .show(5)
```

## DF.where vs spark.sql('SELECT... WHERE')

```python
# where
print("DF.select.where")
df.select("country","city","zipcode","state") \
  .where("state == 'AZ'") \
  .show(5)

df.createOrReplaceTempView("Zipcodes")
print("spark.sql(select where)")
spark.sql(""" SELECT  country, city, zipcode, state FROM ZIPCODES 
          WHERE state = 'AZ' """) \
     .show(5)
```

## DF.orderBy vs spark.sql('SELECT... ORDER BY')

```python
# sorting
print("DF.select.where.orderBy")
df.select("country","city","zipcode","state") \
  .where("state in ('PR','AZ','FL')") \
  .orderBy("state") \
  .show(10)

df.createOrReplaceTempView("Zipcodes")
print("spark.sql(select where order by)")
spark.sql(""" SELECT  country, city, zipcode, state FROM ZIPCODES 
          WHERE state in ('PR','AZ','FL') order by state """) \
     .show(10)
```

## DF.groupBy vs spark.sql('SELECT... GROUP BY')

```python
# grouping
print("DF.groupBy")
df.groupBy("state").count() \
  .show()

df.createOrReplaceTempView("Zipcodes")
print("spark.sql(select group by)")
spark.sql(""" SELECT state, count(*) as count FROM ZIPCODES 
          GROUP BY state""") \
     .show()
```

## Aggregation functions: DF.agg_function

### Core Aggregate Functions
- count(col): Compute the number of items in a group.
- sum(col): Compute the sum of all values in a numeric column.
- avg(col): Compute the average value of a group.
- mean(col): Compute the mean value of a group.
- min(col): Find the minimum values within a group.
- max(col): Find the minimum and maximum values within a group.
- first(col): Retrieve the first value seen in a group (can be non-deterministic without ordering).
- last(col): Retrieve the last value seen in a group (can be non-deterministic without ordering).

```python
from pyspark.sql import functions as F
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("DeptExamples").getOrCreate()

# Employee data: (emp_id, name, salary, dept_id)
emp_data = [(1, "Alice", 70000, 10), 
            (2, "Bob", 50000, 20),
            (3, "Charlie", 80000, 10), 
            (4, "David", 60000, 30),
            (5, "Eve", 75000, 20),
            (6, "Frank", 90000, 10)]

# Department data: (dept_id, dept_name, location)
# dept_data = [(10, "Engineering"), (20, "HR"), (30, "Marketing")]
dept_data = [
    {"dept_id": 10, "dept_name": "Finance", "location": "New York"},
    {"dept_id": 20, "dept_name": "Engineering", "location": "San Francisco"},
    {"dept_id": 30, "dept_name": "HR", "location": "Chicago"}
]

emp_df = spark.createDataFrame(emp_data, ["emp_id", "emp_name", "salary", "dept_id"])
dept_df = spark.createDataFrame(dept_data, ["dept_id", "dept_name"])
joined_df = emp_df.join(dept_df, "dept_id", "inner")  #joining DF, column, type_type
joined_df.show()

# Example of aggregation functions
emp_df.groupBy("dept_id").agg(
    F.count("emp_id").alias("employee_count"),
    F.sum("salary").alias("total_salary"),
    F.avg("salary").alias("avg_salary"),
    F.mean("salary").alias("mean_salary"),
    F.min("salary").alias("min_salary"),
    F.max("salary").alias("max_salary"),
    F.first("salary").alias("first_salary"),
    F.last("salary").alias("last_salary")
).show()
```

## Aggregation functions: spark.sql('SELECT... agg_function')

### Core Aggregate Functions
- count(col): Compute the number of items in a group.
- sum(col): Compute the sum of all values in a numeric column.
- avg(col): Compute the average value of a group.
- mean(col): Compute the mean value of a group.
- min(col): Find the minimum values within a group.
- max(col): Find the minimum and maximum values within a group.
- first(col): Retrieve the first value seen in a group (can be non-deterministic without ordering).
- last(col): Retrieve the last value seen in a group (can be non-deterministic without ordering).

```python
# Must run the above cell first
dept_df.createOrReplaceTempView("departments")
emp_df.createOrReplaceTempView("employees")

spark.sql("""
SELECT e.dept_id, e.emp_id, e.emp_name, e.salary, d.dept_name, d.location
FROM employees e
INNER JOIN departments d ON e.dept_id = d.dept_id
""") \
.show()

# Example of aggregation functions
agg_query = """
SELECT dept_id, 
COUNT(emp_id) as employee_count, 
SUM(salary) as total_salary, 
AVG(salary) as avg_salary,
MEAN(salary) as mean_salary,
MIN(salary) as min_salary,
MAX(salary) as max_salary,
FIRST(salary) as first_salary,
LAST(salary) as last_salary
FROM employees
GROUP BY dept_id
"""
spark.sql(agg_query).show()
```
