```python
from pyspark import SparkContext
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SparkSQL").getOrCreate()
sc = spark.sparkContext

def longer(a,b):
    if (a>b):
        return a
    else:
        return b

def shorter(a,b):
    if (b>a):
        return a
    else:
        return b

# Read file into RDD
lines = sc.textFile("Book.txt")

# map the input lines to line lengths
lineLenghMap = lines.map(lambda x: len(x.split()))

# Find the longest line that has most words.
# Two lines are passed to longer/shorter function which return longer/shorter line
# The process is continue until there is only one line returned.
longest = lineLenghMap.reduce(longer)
shortest = lineLenghMap.reduce(lambda a, b: a if b > a else b) # similar to shorter function but in lambda format
print ("The  most number of words on a line: ", longest)
print ("The least number of words on a line: ", shortest)

```
