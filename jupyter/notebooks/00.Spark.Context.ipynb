## Spark Context

SparkContext (Legacy) is the entry point for any Spark application before Spark 2.0. It represents the connection to a Spark cluster and is responsible for coordinating and managing the resources required for executing tasks. SparkContext helps in setting up the environment and is a gateway to interact with Spark’s low-level API.

Reference: https://spark.apache.org/docs/latest/api/python/reference/api/pyspark.SparkContext.html


```python
# Example of creating a SparkContext 

from pyspark import SparkContext

# Initialize SparkContext
sc = SparkContext("local", "SparkContext Example")

# Create an RDD
rdd = sc.parallelize([1, 2, 3, 4, 5])
print(rdd.collect())

# Stop SparkContext
sc.stop()
```

## Spark Session

Introduced in Spark 2.0, SparkSession acts as a wrapper that combines multiple legacy contexts (SQLContext, HiveContext, and SparkContext) into a single object, simplifying how you interact with different Spark features.

Use SparkSession.builder to create a SparkSession. 

Reference: https://spark.apache.org/docs/latest/api/python/reference/pyspark.sql/spark_session.html

```python
# Example of creating SparkSession and get SparkContext from SparkSession

from pyspark import SparkContext
from pyspark.sql import SparkSession

# Create SparkSesion with SparkSession.builder.  App name is optional but en
spark = SparkSession.builder.appName("SparkSession and Context").getOrCreate()

# Get SparkContext from SparkSession
sc = spark.sparkContext

# Create an RDD
rdd = sc.parallelize([1, 2, 3, 4, 5])
print(rdd.collect())

# Stop SparkContext
sc.stop()
```

```python
# Example of creating SparkSession and use it to read csv file to a Spark DataFrame

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SparkSession and Context").getOrCreate()

# Create DataFrame
df = spark.read \
          .option("header",True) \
          .csv("simple-zipcodes.csv")

print("DF.printSchema")
df.printSchema()

print("DF.show")
df.show()
```
