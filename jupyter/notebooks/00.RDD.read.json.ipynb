```python
import json
from pyspark import SparkContext
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SparkSQL").getOrCreate()
sc = spark.sparkContext

# Read input
lines = sc.textFile("simple-zipcodes.json")
jsons = lines.map(lambda x: json.loads(x)) # load each string to json (a dictionary)

# Print the input text lines
print("Lines on the input text file:")
results = lines.collect()
for result in results:
    print(result) # a string
print()

# Print jsons
print(f"jsons: {jsons}")
results = jsons.collect()
for result in results:
    print(result) # a dictionary
```
