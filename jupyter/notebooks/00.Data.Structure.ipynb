## Sequence

A sequence is a data type that adheres to the "sequence protocol", meaning its elements are ordered and can be accessed using an integer index (including negative indices for counting from the end) and slicing. Common operations applicable to all sequences include: 
- Indexing: Accessing individual elements (e.g., my_sequence[0]).
- Slicing: Extracting a portion of the sequence (e.g., my_sequence[1:3]).
- Length: Using the len() function.
- Iteration: Looping through elements in order.
- Membership testing: Using the in and not in operators. 

Examples of built-in Python sequence types are: 
- list
- tuple
- str (strings)
- range (a function to generate a range of sequence which can used to form a list or a loop)
- bytes and bytearray

Non built-in Python sequence types are: 
- array
- numpy.narray

## Tuple
In Python, a tuple is an ordered, immutable sequence of elements. This means that once a tuple is created, its contents cannot be changed (elements cannot be added, removed, or modified). Tuples are defined by enclosing comma-separated values within parentheses ().

## Range function

The range() function in Python is used to generate a sequence of integers within a specified range. It is most commonly used in loops to control how many times a block of code runs.

Syntax: range(start, stop, step)
Default: start = 0, step = 1
Valid values: start, stop and step can be negative or positive.

```python
# Default start and step
for i in range(2):
    print(i, end=" ")
print()

# start, stop and step can be negative
for i in range(-3,-5,-1):
    print(i, end=" ")
print()

# Using range in List Comprehension
list1 = [x for x in range(10) if x % 2 == 0]
print(list1)
```

## Enumeration on List

The enumerate() function in Python is a built-in function that adds a counter to an iterable object and returns an enumerate object. This object yields pairs of (index, value) for each item in the iterable, making it convenient to access both the position and the element during iteration.

## The map() function

The Python map() function applies a given function to each item of an iterable (like a list, tuple, or set).
It returns a map object (an iterator) containing the results. 
This allows for efficient data transformation without using explicit loops.  

Syntax: map(function, iterable, ...)
* function: The function to execute for each item.  The function can be a lambda.
* iterable: One or more sequence, collection, or iterator objects. The function must accept as many arguments as there are iterables.
* Return: In Python 3, map() returns a lazy iterator (a map object) that yields results on demand, which is memory efficient for large datasets. To see the results immediately as a list, you must convert the map object using list() or a star operator (unpack operator)


```python
# Example using tradictional function
def square(number):
  return number ** 2

numbers = [1, 2, 3, 4, 5]
squared_numbers_map = map(square, numbers)
print(f"map function returns type: {type(squared_numbers_map)}")

# Convert the map object to a list to view results
# squared_numbers_list = list(squared_numbers_map) # runtime error: 'list' object is not callable
squared_numbers_list = [*squared_numbers_map]
print(squared_numbers_list)
```

```python
# Example using lambda

celsius_temps = [0.0, 10.0, 20.0, 30.0, 40.0]
# Convert to Fahrenheit: (temp * 9/5) + 32
fahrenheit_temps = [*(map(lambda temp: (temp * 9/5) + 32, celsius_temps))] # Create a list after * which unpacking the iterator
print(fahrenheit_temps)
```

```python
en_stopwords = ['a', 'the']
text_list = ["a is good", "b is the mean"]
text_map = map(lambda x: ' '.join([word for word in x.split() if word not in (en_stopwords)]), text_list)
clean_text_list = [*text_map]
print (clean_text_list)
```

## Queue
REF: https://realpython.com/ref/stdlib/queue/

Python provides several built-in ways to implement queues, each suited for different use cases.

Queues are linear data structures used to manage data or tasks in a "First In, First Out" (FIFO) manner, ensuring the first element added is the first one removed. They act as buffers to hold data for later processing, manage asynchronous communication between systems, and handle task scheduling to optimize resource usage. 

Common Uses of Queues:
- Operating Systems and Task Scheduling: Used for CPU scheduling, job scheduling for printers, and managing I/O requests where tasks wait to be processed.
- Asynchronous Processing (Message Queues): Used in software architecture to decouple services, allowing a system to handle traffic spikes, manage background tasks, and process tasks "fire-and-forget" style.
- Data Structures and Algorithms: Implemented for breadth-first search (BFS) on graphs and managing queues for undo functionality in applications.
- Real-world Modeling: Applied in scenarios requiring ordered servicing, such as call center, call handling, and customer ticketing systems.

Examples of Queue Use Cases:
- Web Servers: Handling incoming requests in order.
- Print Spooling: Managing documents sent to a printer in the order they were submitted.
- Data Buffering: Handling data streaming or high-volume data transfers.
- Message Queues: Technologies like RabbitMQ or Kafka used for service communication.


```python
# Using the queue Module (for Threading) 

import queue

# Create a thread-safe queue with a maximum size of 3
q = queue.Queue(maxsize=3)

# Add items using put()
q.put('Task A')
q.put('Task B')
q.put('Task C')

# Remove items using get()
item1 = q.get()
item2 = q.get()

print(f"Processed {item1}")
print(f"Processed {item2}")
print(f"Queue size after getting items: {q.qsize()}")
```

```python
# Basic Usage with collections.deque for typical single-threaded use.
from collections import deque

# Create a queue
my_queue = deque()

# Add items (enqueue)
my_queue.append('Task 1')
my_queue.append('Task 2')
my_queue.append('Task 3')

print(f"Current queue: {list(my_queue)}")  # create a list from a queue

# Remove items (dequeue)
first_item = my_queue.popleft()
second_item = my_queue.popleft()

print(f"Removed item: {first_item}")
print(f"Removed item: {second_item}")
print(f"Remaining queue: {list(my_queue)}")
```

## Stack

A stack is a linear data structure that operates on the Last-In, First-Out (LIFO) principle, meaning the most recently added element is the first one to be removed. Stacks are essential in computer science for managing data that needs to be accessed in reverse order, such as tracking recent actions, managing function calls, or parsing nested structures. 

Here are the primary uses of stacks:
1. Function Call Management (Call Stack) 
- Recursion: When a function calls itself, the system uses a stack to store a "snapshot" of each call, including local variables, so it can return to the correct place when the function finishes.
- Procedure Calls: In programming, when one function calls another, the current execution state is "pushed" onto the stack; when the function finishes, it is "popped" to return control to the caller. 

2. "Undo" Mechanisms in Applications
- Undo/Redo: Text editors (like Notepad), graphic editors (like Photoshop), and browsers use a stack to track actions. The last action performed is pushed onto the stack. When you hit "Undo," that last action is popped and reversed.
- Browser History: Every URL you visit is pushed onto a stack. Clicking the "Back" button pops the most recent page to take you to the previous one. 

3. Syntax Parsing and Validation
- Balanced Brackets: Compilers and IDEs use stacks to verify if code is correctly formed by matching opening symbols (e.g., (, [, {) with corresponding closing symbols (e.g., ), ], }).
- Expression Evaluation: Calculators use stacks to handle operator precedence (order of operations) and convert infix expressions (e.g., 2+3) to postfix or prefix notation. 

4. Backtracking Algorithms
- Maze Solving & Graph Traversal: When searching for a path, you can use a stack to push the current path. If you hit a dead end, you can pop from the stack to "backtrack" to the last known correct point.
- Depth-First Search (DFS): Algorithms that traverse trees or graphs, such as finding a path in a maze, use a stack to remember which node to visit next. 

5. Memory Management
- Automatic Variable Allocation: A specific region of RAM, known as the "stack," is used for fast, automatic allocation and deallocation of memory for local variables within functions.

```python
# Implement a stack
class Stack:
    def __init__(self):
        self._items = []

    def push(self, item):
        self._items.append(item)

    def pop(self):
        if self.is_empty():
            raise IndexError("pop from empty stack")
        return self._items.pop()

    def peek(self):
        if self.is_empty():
            raise IndexError("peek from empty stack")
        return self._items[-1]

    def is_empty(self):
        return len(self._items) == 0

    def size(self):
        return len(self._items)


s = Stack()
s.push(10)
s.push(20)
print(s.pop())   # 20
print(s.peek())  # 10
print(s.size())
```
