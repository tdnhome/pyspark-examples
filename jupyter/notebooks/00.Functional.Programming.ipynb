***Functional Programming***

Functional programming (FP) ***in general*** is a declarative paradigm that ***constructs software by composing pure functions and avoiding shared state, mutable data, and side effects***. By emphasizing immutability and predictable, independent functions, it enhances code maintainability, testing, and easy parallelization. It treats functions as first-class citizens.

Functional programming (FP) ***in Python*** is about writing code using pure functions, immutability, and composition — treating computation as the evaluation of functions rather than changing state.

***Pure Function***

A pure function is a deterministic function that always returns the same output for the same input and produces no side effects, such as modifying external variables, interacting with APIs, or updating the DOM. It relies solely on its input arguments, making the code easier to test, debug, and reuse. 

Key Characteristics of Pure Functions:
- ***Predictable Output***: Given the same arguments, a pure function will always return the same value, regardless of how many times it is called or the application state.
- ***No Side Effects***: A pure function does not alter any state outside its scope. It does not modify input data (treating data as immutable), write to files, log to the console, or make network requests.
- ***Dependence on Inputs***: The result is calculated exclusively from the arguments passed to the function

Python isn’t purely functional, but it ***provides powerful functional programming tools*** that are widely used in real systems.

## Core Idea of Functional Programming and High-Order Function

Functional programming (FP) is a programming paradigm or style, while a higher-order function (HOF) is a specific type of function used within that paradigm. In essence, HOFs are a key tool that enables functional programming.
Higher-order functions are a powerful mechanism within Python that facilitate adopting a functional programming style. 

<b>Functional Programming:</b> Functional programming is an overall approach to writing code that emphasizes the use of mathematical functions and immutability. Build programs by combining small, reusable functions instead of mutating data.

Key principles include: 
- Pure Functions: Functions should produce the same output for the same input and have no side effects (e.g., modifying external state or variables).
- Immutability: Data, once created, generally cannot be changed.
- Function Composition: Building complex logic by combining simpler functions into processing pipelines.
- First-Class Functions: Functions are treated as values that can be assigned to variables, passed as arguments, and returned from other functions. This is a prerequisite for using HOFs

<b>Higher-Order Function:</b> A higher-order function is a function that does at least one of the following: 
- Takes one or more functions as arguments.
- Returns a function as its result. 

In Python, popular built-in examples of HOFs include: 
- map()
- filter()
- functools.reduce()




## map()

The built-in map() function in Python is a high-order, functional tool that applies a given function to every item in an iterable (like a list or tuple) and returns an iterator with the results. It promotes cleaner, faster code by replacing explicit for loops for data transformations. 

Key Aspects of map()
- Syntax: map(function, iterable, ...).
- Parameters: Takes a function (or lambda) as its first argument and one or more iterables as the second argument.
- Return Value: Returns a map object, which is an iterator. This iterator yields transformed items only when required, improving efficiency.

Using multiple iterables: If multiple iterables are provided, map stops when the shortest one is exhausted.   

map() is highly efficient because it performs transformations lazily, making it a staple for functional programming in Python


```python
# Applying a lambda function, x=x*2, to a list
m = map(lambda x: x * 2, [1, 2, 3])
print(list(m)) # convert the map to list

# Converts numbers to strings using str function
print(list(map(str, [1, 2, 3]))) # str is a function
```

## filter()

The built-in filter() function in Python constructs an iterator from elements of an iterable (list, tuple, etc.) for which a specified function returns true. It acts as a functional tool to extract subsets of data based on a condition, yielding only the elements where the function is truthy. 

Syntax:
filter(function, iterable)
- function: Checks if each element is true or false. Can be None to remove falsy values (e.g., None, 
, False).
- iterable: The collection to filter (e.g., list, set, generator). 

Key Details
- Return Value: A filter object (an iterator), which can be converted to a list() or tuple() to view immediately.
- Usage: It provides a cleaner, functional alternative to explicit for-loops for selecting data.

Alternatives:
- Often, filter() is replaced by list comprehensions (e.g., [n for n in numbers if n > 0]) or generator expressions, which are frequently considered more readable
- See List-Comprehension module


```python
# Filter positive numbers
numbers = [-2, -1, 0, 1, 2]
positive_numbers = filter(lambda n: n > 0, numbers)
print(list(positive_numbers))

# Filer using List Comprehensions
p = [n for n in numbers if n > 0]
print(p)
```

## reduce - from functools library

The functools.reduce() function in Python applies a function cumulatively to the items of an iterable, from left to right, to reduce it to a single final value.

<b>Usage:</b> To use reduce(), you must first import it from the functools module. 

<b>Syntax:</b> reduce(function, iterable[, initializer])
- function: A binary function (takes two arguments) that performs the cumulative operation.
- iterable: The sequence of values (e.g., list, tuple, etc.) to be processed
- initializer (optional): A starting value that is placed before the items of the iterable in the calculation. If not provided, the first two elements of the iterable are used to start the process. 

Note: the iterable can also be a list of functions, which then become a composition of function.  See next section

```python
# To calculate the sum of all numbers in a list:
from functools import reduce

numbers = [1, 2, 3, 4, 5]
total = reduce(lambda x, y: x + y, numbers)
print(total)
```

## Function Composition

Composing multiple functions with functools.reduce.  This is a data pipe line job

```python
from functools import reduce

def compose_all(*functions):
    """Composes an arbitrary number of functions from right to left."""
    def apply(x, f):
        return f(x)
    # reduce applies the 'apply' function sequentially across the 'functions' list
    # The initial value for the reduction is the input 'x'
    return lambda x: reduce(apply, functions, x)

# Define a list of functions to be in the composition:
def square(x):
    return x*x

def inc(x):
    return x+1

def double(x):
    return x*2

# Composes functions from left to right (square -> inc -> double)
convert = compose_all(square, inc, double)
result = convert(3)
print(result)
# Output: 3*3=9+1=10*2=20
```

## FP vs Imperative Example

Below are examples of imprative, FP, and Pythonic FP

Given a list of numbers, select even all even numbers.  For each element, double the value. 

```python
# Imperative
nums = [1, 2, 3, 4, 5]
result = []
for x in nums:
    if x % 2 == 0:
        result.append(x * 2)
print(result)
```

```python
# Functional
nums = [1, 2, 3, 4, 5]
result = list(map(lambda x: x*2, filter(lambda x: x%2==0, nums)))
print(result)
```

```python
# Pythonic FP - list comprehension (best)
nums = [1, 2, 3, 4, 5]
result = [x * 2 for x in nums if x % 2 == 0]
print(result)
```
